# Header Pattern Exploration

## Objective

Explore section header patterns in ICU notes and empirically determine common headers for rule-based section segmentation.

```python
import pandas as pd
import re
from collections import Counter
```

```python
CORPUS_PATH = "../../data/processed/icu_corpus.csv"
TEXT_COLUMN = "TEXT"

df = pd.read_csv(CORPUS_PATH)
notes = df[TEXT_COLUMN].dropna().tolist()

print(len(notes), "notes loaded")
```

```python
header_counter = Counter()

colon_pattern = re.compile(r"^\s*([A-Za-z][A-Za-z0-9 /()\-'&]{0,80})\s*:\s*")
standalone_pattern = re.compile(r"^\s*([A-Za-z][A-Za-z0-9 /()\-\']{1,80})\s*:?\s*$")

for note in notes:

    lines = note.split("\n")

    for line in lines:

        match = colon_pattern.match(line)

        if match:
            header_counter[match.group(1)] += 1
            continue

        match = standalone_pattern.match(line)

        if match:
            header_counter[match.group(1)] += 1


for header, count in header_counter.most_common(300):
    print(f"{count:6}  {header}")
```
